```python
# JDME Procesamiento de lenguaje natural
from transformers import pipeline
```

```python
clasificador = pipeline(
    "sentiment-analysis",
    model="dccuchile/bert-base-spanish-wwm-cased",
    tokenizer="dccuchile/bert-base-spanish-wwm-cased",
)
```

```python
textos = [
    "Me encanta este producto, superó todas mis expectativas",
    "El servico al cliente fue pésimo y el envio tardó demasiado",
    "La calidad es decente, pero no es nada del otro mundo",
    "Era justo lo que buscaba",
    "Todos deberíamos tener uno"
]
```

```python
prediccion = clasificador(textos)
```

```python
for texto, resultado in zip(textos, prediccion):
    sentimiento = resultado["label"]
    puntuacion = resultado["score"]
    print("-" * 30)
    print(f"Texto: ***{texto}***")
    print(f"Sentimiento: ***{sentimiento}***")
    print(f"Confianza: ***{puntuacion:.4f}***")
```
